4.2. Linear Regression

```python
# Code source: Jaques Grobler
# License: BSD 3 clause
# =============================================================================
# Linear Regression Example
# =============================================================================
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.model_selection import train_test_split

# Load the diabetes dataset
diabetes = datasets.load_diabetes()
X = diabetes.data
y = diabetes.target

# Split the data into training/testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(X_train, y_train)

# Make predictions using the testing set
y_pred = regr.predict(X_test)

# The coefficients
print('Coefficients:\n', regr.coef_)
# Intercept
print('Intercept:\n', regr.intercept_)
# Explained variance score: 1 is perfect prediction
print("R^2 = %0.5f" % r2_score(y_test, y_pred))
# The mean absolute error
print("MAE = %5.3f" % mean_absolute_error(y_test, y_pred))
# The mean squared error
print("MSE = %5.3f" % mean_squared_error(y_test, y_pred))

# Plot outputs
fig, ax = plt.subplots()
ax.scatter(y_test, y_pred, edgecolors=(0, 0, 0))
ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```

4.3. Ridge Regression

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.model_selection import train_test_split

# Load the diabetes dataset
diabetes = datasets.load_diabetes()
X = diabetes.data  # Assign dataset features
y = diabetes.target  # Assign target variable

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)

# Create Ridge regression model with alpha=0.5
regr = linear_model.Ridge(alpha=0.5)

# Train the model
regr.fit(X_train, y_train)

# Make predictions using the test set
y_pred = regr.predict(X_test)

# Print model coefficients
print('Coefficients:\n', regr.coef_)
print('Intercept:\n', regr.intercept_)

# Evaluate performance metrics
print("R^2 = %0.5f" % r2_score(y_test, y_pred))
print("MAE = %5.3f" % mean_absolute_error(y_test, y_pred))
print("MSE = %5.3f" % mean_squared_error(y_test, y_pred))

# Plot outputs
fig, ax = plt.subplots()
ax.scatter(y_test, y_pred, edgecolors=(0, 0, 0))
ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```

4.4. Lasso Regression

```python
# =============================================================================
# Lasso Regression Example
# =============================================================================

# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.model_selection import train_test_split

# Load the California housing dataset (since Boston dataset is deprecated)
housing = datasets.fetch_california_housing()
X = housing.data  # Features
y = housing.target  # Target variable

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)

# Create Lasso regression model with alpha=0.1
regr = linear_model.Lasso(alpha=0.1)

# Train the model
regr.fit(X_train, y_train)

# Make predictions using the test set
y_pred = regr.predict(X_test)

# Print model coefficients
print('Coefficients:\n', regr.coef_)
print('Intercept:\n', regr.intercept_)

# Evaluate performance metrics
print("R^2 = %0.5f" % r2_score(y_test, y_pred))
print("MAE = %5.3f" % mean_absolute_error(y_test, y_pred))
print("MSE = %5.3f" % mean_squared_error(y_test, y_pred))

# Plot outputs
fig, ax = plt.subplots()
ax.scatter(y_test, y_pred, edgecolors=(0, 0, 0))
ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```

5. TASK

1. Similarities
- All three are regression models: They predict a continuous output variable (like house prices or medical test results).
-  Based on the same equation: Each model follows the linear equation:y=0
 +1.X1 + 2.X2 + ...+n.Xn
- Use the Least Squares method: They minimize errors in predictions.
- Work with multiple features: All three methods handle datasets with many input variables.
- Measure performance using similar metrics: R Score, Mean Absolute Error (MAE), and Mean Squared Error (MSE) are commonly used to evaluate them.

2. Difference:

- Regularization: Linear Regression does not apply any penalty term, while Ridge Regression uses L2 regularization to penalize large coefficients, and Lasso Regression applies L1 regularization, which can shrink some coefficients to zero.

- Overfitting Prevention: Linear Regression can overfit when there are too many features, whereas Ridge and Lasso help reduce overfitting.

- Feature Selection: Linear and Ridge Regression use all features, but Ridge only reduces their impact. Lasso, however, can remove unimportant features entirely by shrinking coefficients to zero.

- Effect on Coefficients: Linear Regression keeps all coefficients, Ridge reduces their values without making them zero, and Lasso can force some to become exactly zero.

- Best Used When: Linear Regression is suitable for small datasets without multicollinearity. Ridge is better for large datasets with highly correlated features, and Lasso is ideal when feature selection is needed.

- Computational Complexity: Linear Regression is the simplest, while Ridge and Lasso are more computationally complex


3. Which One is Better?

If the dataset is small and clean  Linear Regression works well.

If the dataset has highly correlated features (multicollinearity)  Ridge Regression is better.

If you want to select only the most important features  Lasso Regression is the best choice.
